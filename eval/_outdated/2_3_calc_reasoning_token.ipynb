```python
import tiktoken
import pandas as pd

# 토크나이저 설정
encoding = tiktoken.get_encoding("o200k_base")

# CSV 파일 읽기
df = pd.read_csv("./drive/MyDrive/Colab Notebooks/독자AI/추론/12번_fn/(FN)12_70건_전처리 및 셔플 완료.csv_gpt-oss-120b-reasoning.csv", encoding="utf-8")
# A, B, C, D, E 컬럼의 각 셀을 토큰 수로 변환
columns_to_convert = [
    'shuffled_balance_text',
    'shuffled_cluster_front_text',
    'shuffled_cluster_middle_text',
    'shuffled_cluster_end_text',
    'shuffled_random_text'
    ]

for col in columns_to_convert:
    df[col] = df[col].apply(lambda x: len(encoding.encode(str(x))) if pd.notna(x) else 0)

# 결과 확인
print(f"변환 완료: {len(df)}행 × {len(columns_to_convert)}컬럼 = {len(df) * len(columns_to_convert)}개")
print("\n처음 5행:")
print(df[columns_to_convert].head())

print("\n통계:")
print(f"평균 토큰 수: {df[columns_to_convert].values.mean():.2f}")
print(f"총 토큰 수: {df[columns_to_convert].values.sum()}")

# 결과 저장
df.to_csv("./drive/MyDrive/Colab Notebooks/독자AI/추론/token_counts.csv", index=False, encoding="utf-8-sig")
print("\n✓ token_counts.csv 저장 완료")
```
